```python
#PHSX 216 Projectile Motion Lab plotting code
#performs 2nd degree polynomial fit
#Brianna Beller
#Updated September 2024

import numpy as np
import matplotlib.pyplot as plt

#-----------------------------------------------------------------------#
#----------UPDATE THIS SECTION!!!----------

#DATA TO BE PLOTTED
#UPDATE THE VARIABLE NAMES & NUMBERS (LEAVE BRACKETS/PARENTHESES AS IS)
#variable names should start with a letter and must contain only letters, numbers, and underscores

xvariable_changeme = np.array([36, 39, 42, 45, 48]) #what are units?   <-----

yvariable_changeme = np.array([0.625, .640, .650, .654, .501]) #what are units?   <-----

err_yvariable_changeme = np.array([.011, .010, .016, .012, .024])                   #                 <-----

#Re-assign variables as x, y, dy so that the remaining plotting code may remain generic
#UPDATE VARIABLE NAMES (ON RHS OF EACH EQUAL SIGN) TO MATCH VARIABLE NAMES FOR DATA ABOVE

x = xvariable_changeme             #this should be the array you want to plot on the x axis #        <-----
y = yvariable_changeme             #this should be the array you want to plot on the y axis #        <-----
dy = err_yvariable_changeme        #this should be your error in y array                    #        <-----

#UPDATE EACH OF THESE 3 STRINGS FOR YOUR PLOT LABELS AS YOU WANT THEM TO APPEAR

plot_title= 'Graph of mean y value versus theta'                                                                #                 <-----
x_label= ' (degrees)'                                                    #                 <-----
y_label= 'Mean y value (meters)'                                               #                 <-----

#-----------------------------------------------------------------------#
#-----DON'T NEED TO CHANGE ANYTHING BEYOND THIS POINT!!!-----


#size the plot
plt.figure(figsize=(15,10))

#create scatter plot
plt.scatter(x, y, color='blue', marker='o')

#create labels
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(plot_title)

#fitting to a 2nd degree polynomial
c,b,a=np.polynomial.polynomial.polyfit(x,y,2,w=dy)

#Annotate with values of A, B, C from best fit polynomial
plt.annotate('best fit line: $f(x)=A^2+B+C$',(0.05, 0.5), xycoords='axes fraction')

plt.annotate('     A = {:.6f}'.format(a),(0.05, 0.45), xycoords='axes fraction')

plt.annotate('     B = {:.6f}'.format(b),(0.05, 0.40), xycoords='axes fraction')

plt.annotate('     C = {:.6f}'.format(c),(0.05, 0.35), xycoords='axes fraction')

#Create fit line
xnew = np.linspace(x.min(), x.max(), 300)
fit = a*xnew**2 + b*xnew +c

plt.plot(xnew, fit, color='green')
plt.show()
```
